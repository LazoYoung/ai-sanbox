```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

x = np.linspace(0, 6, 100)
y = np.linspace(-1, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) - np.cos(Y)
# z = x**2 + y**2

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_aspect('equal')
plt.show()
```
